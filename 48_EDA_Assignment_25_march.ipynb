### Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.


```python
dataset = df
app_names = dataset['App']
installs = dataset['Installs']

installs = installs.str.replace(',', '').str.replace('+', '').astype(int)

df = pd.DataFrame({'App': app_names, 'Installs': installs})

app_installs = df.groupby('App')['Installs'].sum()

top_10_apps = app_installs.nlargest(10)

frequency_table = pd.DataFrame({'App': top_10_apps.index, 'Number of Installs': top_10_apps.values})

frequency_table.reset_index(drop=True, inplace=True)
print(frequency_table)
```
