```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
column_names = ["Class", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium",
                "Total phenols", "Flavanoids", "Nonflavanoid phenols", "Proanthocyanins",
                "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"]
```

```python
df = pd.read_csv('wine.data',names=column_names)
```

```python
df.head()
```

```python
X = df.drop('Class',axis=1)
y =df['Class']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
X_pca = pca.fit_transform(x_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
```

```python
n_components = np.argmax(cumulative_variance >= 0.95) + 1
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Scatter Plot')
plt.colorbar(label='Class')
plt.show()
```

```python
from sklearn.cluster import KMeans 
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X_pca)
```

```python
df['Cluster'] = y_kmeans
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(X_pca, y_kmeans)
```

```python
print(f"Number of retained principal components: {n_components}")
print(f"Silhouette Score: {silhouette_avg}")
print("\nCluster Centers:")
cluster_centers_df = pd.DataFrame(pca.inverse_transform(kmeans.cluster_centers_), columns=X.columns)
print(cluster_centers_df)
```
